```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

import pandas as pd

csv_path = "/content/drive/MyDrive/archive (3)/COVID clinical trials.csv"
df = pd.read_csv(csv_path)

print("Trials in CSV:", len(df))
df.head()

```

```python
import os
from lxml import etree

xml_dir = "/content/drive/MyDrive/COVID-19 CLinical trials studies/COVID-19 CLinical trials studies"

def extract_inclusion_exclusion(xml_path):
    try:
        tree = etree.parse(xml_path)
        root = tree.getroot()

        textblock = root.xpath(".//eligibility/criteria/textblock")
        if not textblock:
            return None, None

        text = textblock[0].text or ""
        text_lower = text.lower()

        if "exclusion" in text_lower:
            split_idx = text_lower.index("exclusion")
            inclusion = text[:split_idx].strip()
            exclusion = text[split_idx:].strip()
        else:
            inclusion = text.strip()
            exclusion = ""

        return inclusion, exclusion
    except:
        return None, None
```

```python
eligibility_records = []

for file in os.listdir(xml_dir):
    if file.endswith(".xml"):
        nct_id = file.replace(".xml", "")
        inc, exc = extract_inclusion_exclusion(
            os.path.join(xml_dir, file)
        )
        eligibility_records.append({
            "NCT Number": nct_id,
            "Inclusion Criteria": inc,
            "Exclusion Criteria": exc
        })

elig_df = pd.DataFrame(eligibility_records)
print("Trials with eligibility:", len(elig_df))
elig_df.head()
```

```python
df = df.merge(elig_df, on="NCT Number", how="inner")

df = df[
    df["Inclusion Criteria"].notna() &
    df["Exclusion Criteria"].notna()
].reset_index(drop=True)

print("Final usable trials:", len(df))
```

```python
df["inclusion_text"] = (
    "Trial ID: " + df["NCT Number"] +
    "\nTitle: " + df["Title"] +
    "\nInclusion Criteria:\n" + df["Inclusion Criteria"]
)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(
    df["inclusion_text"].tolist(),
    show_progress_bar=True
)
```

```python
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))

print("FAISS index size:", index.ntotal)
```

```python
patient_profile = """
Age: 55
Gender: Male
Confirmed COVID-19
Hospitalized
Comorbidity: Diabetes
No kidney disease
Not pregnant
"""
```

```python
TOP_K = 10
patient_embedding = model.encode([patient_profile])

distances, indices = index.search(patient_embedding, TOP_K)
candidate_trials = df.iloc[indices[0]]

candidate_trials[["NCT Number", "Title"]]
```

```python
def build_eligibility_prompt(patient_profile, inclusion_criteria, exclusion_criteria):
    return f"""
Patient Profile:
{patient_profile}

Inclusion Criteria:
{inclusion_criteria}

Exclusion Criteria:
{exclusion_criteria}

Question:
Is the patient eligible for this clinical trial?

Answer STRICTLY in one of the following forms:
ELIGIBLE - <short reason>
NOT ELIGIBLE - <short reason>
"""
```

```python
def rule_based_eligibility(patient_profile, inclusion, exclusion):
    patient = patient_profile.lower()
    inclusion = inclusion.lower()
    exclusion = exclusion.lower()

    # Hard exclusions
    if "kidney" in patient and "kidney" in exclusion:
        return "NOT ELIGIBLE", ["Kidney disease is listed as an exclusion criterion"]

    if "pregnant" in patient and "pregnant" in exclusion:
        return "NOT ELIGIBLE", ["Pregnancy is listed as an exclusion criterion"]

    # Inclusion checks
    reasons = []

    if "covid" in patient and "covid" in inclusion:
        reasons.append("Patient has COVID-19 and the trial includes COVID-19 patients")

    if "diabetes" in patient and "diabetes" in inclusion:
        reasons.append("Trial includes patients with diabetes")

    if not reasons:
        reasons.append("Patient broadly matches the inclusion criteria")

    return "ELIGIBLE", reasons
```

```python
final_results = []

for _, row in candidate_trials.iterrows():
    decision, reasons = rule_based_eligibility(
        patient_profile,
        row["Inclusion Criteria"],
        row["Exclusion Criteria"]
    )

    final_results.append({
        "Trial ID": row["NCT Number"],
        "Title": row["Title"],
        "Decision": decision,
        "Justification": reasons
    })
```

```python
for i, r in enumerate(final_results, 1):
    print(f"\n{i}. Trial ID: {r['Trial ID']}")
    print(f"Title: {r['Title']}")
    print(f"Decision: {r['Decision']}")
    print("Justification:")
    for reason in r["Justification"]:
        print(f"- {reason}")
```
